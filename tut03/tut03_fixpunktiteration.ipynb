```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    a = -0.5
    b = 2
    c = 0
    return a*x**2 + b*x + c
```

```python
x = np.linspace(0,4,100)
y = f(x)

x0 = 0.3
erg = np.zeros(10)
erg[0] = x0
for i in range(len(erg)-1):
    erg[i+1] = f(erg[i])

print("Iterationen: \n", erg)
```

```python
# PLOTS

ax = plt.gca()
ax.plot(x,y, label='$f(x) = -0.5x^2 + 2x$') # plot function
ax.plot(x,x, label='$id(x) = x$')
ax.set_xlim((0,4))
ax.set_ylim((0,2.2));

# plot iterations
x_help = [[item, item] for item in erg]

x_print = np.array([item for sublist in x_help for item in sublist])
y_print = np.zeros(len(x_print))
for i in range(int(0.5*len(x_print))):
    y_print[2*i]   = x_print[2*i]
    y_print[2*i+1] = f(x_print[2*i+1])
y_print[0] = 0

ax.plot(x_print,y_print);
ax.legend();
```
